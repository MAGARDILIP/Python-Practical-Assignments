## 1. buffered Cursor

**Purpose:**  
A buffered cursor reads all query results immediately and avoids the "Unread result found" error in MySQL.

**What happens if not used:**  
If `buffered=True` is not used, MySQL may throw an error when closing the cursor because some results are still unread.

**Syntax:**  
connection.cursor(buffered=True)

## 2. DELETE Statement (MySQL)

**Purpose:**  
The DELETE statement is used to remove a specific record from a table based on a condition.

**Syntax:**  
DELETE FROM table_name WHERE condition

---

## 3. Transactions in MySQL

**Purpose:**  
A transaction is used to execute multiple SQL statements as a single unit of work.  
Either all operations succeed (COMMIT) or none are applied (ROLLBACK).

---

## start_transaction()

**Purpose:**  
Used to explicitly start a transaction in MySQL.

**Syntax:**  
connection.start_transaction()

---

## commit() in Transactions

**Purpose:**  
Saves all changes made during the transaction permanently.

**Syntax:**  
connection.commit()

---

## rollback()

**Purpose:**  
Reverts all changes made during the transaction if an error occurs.

**Syntax:**  
connection.rollback()                                                                                                                                          ## try–except Block (with Transactions)

**Purpose:**  
The try–except block is used to handle runtime errors and ensure safe execution of database transactions.

**Why it is used:**  
If an error occurs during a transaction, the except block allows the program to roll back all changes, preventing partial updates.

**Outcome:**  
- try → execute SQL statements  
- commit → if no error  
- rollback → if error occurs

## 4. SUM() Aggregate Function

**Purpose:**  
The SUM() function is used to calculate the total of a numeric expression or column.

**Syntax:**  
SELECT SUM(column_name) FROM table_name

**Example:**  
SELECT SUM(quantity * price) FROM orders WHERE product_name = 'Laptop'

## 5. Prepared Statement (MySQL)

**Purpose:**  
A prepared statement is used to execute the same SQL query multiple times with different values in a safe and efficient way.

**Benefits:**  
- Improves performance  
- Prevents SQL injection  
- Reduces query parsing overhead  



## executemany()

**Purpose:**  
The executemany() method is used to execute a prepared SQL statement multiple times with different values.

**Syntax:**  
cursor.executemany(sql_query, list_of_tuples)

## Prepared UPDATE Statement

**Purpose:**  
A prepared UPDATE statement is used to safely update records in a table using placeholders.

**Benefits:**  
- Prevents SQL injection  
- Improves query safety and readability  

**Syntax:**  
UPDATE table_name SET column_name = %s WHERE condition


---

## 6. FOREIGN KEY (Table Relationship)

**Purpose:**  
A FOREIGN KEY is used to establish a relationship between two tables by linking a column in one table to the primary key of another table.

**Why it is used:**  
- Maintains data integrity  
- Ensures valid references between tables  

**Syntax:**  
FOREIGN KEY (column_name) REFERENCES parent_table(primary_key)

---

## 7. JOIN Operation (MySQL)

**Purpose:**  
The JOIN operation is used to retrieve data from two or more related tables based on a common column.

**Why it is used:**  
- Combines data from related tables  
- Displays meaningful information from relationships  

**Syntax:**  
SELECT columns FROM table1 JOIN table2 ON table1.column = table2.column

**1.Write a Python program that connects to a MySQL database, updates the salary of
an employee in the employees table, and then queries and prints the updated
record**


```python
import mysql.connector

# connect to MySQL database
con = mysql.connector.connect(
    host="localhost",
    user="root",
    password="2910",
    database="college"
)

# buffered cursor is used to avoid unread result error
cur = con.cursor(buffered=True)

# show data BEFORE update
print("Employee data BEFORE update:")
cur.execute("SELECT name, salary FROM employees WHERE name = %s", ("Dilip",))
row = cur.fetchone()
print(row[0], "-", row[1])

# update salary
cur.execute("UPDATE employees SET salary = %s WHERE name = %s", (606006, "Dilip"))

# commit is required to save the update permanently
con.commit()

# show data AFTER update
print("\nEmployee data AFTER update:")
cur.execute("SELECT name, salary FROM employees WHERE name = %s", ("Dilip",))
row = cur.fetchone()
print(row[0], "-", row[1])

# close cursor and connection
cur.close()
con.close()
```

MySQL database, start a transaction, insert multiple records into a table, and
commit the transaction.**


```python
import mysql.connector

# connect to MySQL database
con = mysql.connector.connect(
    host="localhost",
    user="root",
    password="2910",
    database="college"
)

# buffered cursor to avoid unread result error
cur = con.cursor(buffered=True)

try:
    # start transaction
    con.start_transaction()
    print("Transaction started")

    # insert multiple records
    cur.execute("INSERT INTO employees (name, salary) VALUES (%s, %s)", ("Tejas", 48000))
    cur.execute("INSERT INTO employees (name, salary) VALUES (%s, %s)", ("Vicky", 52000))
    cur.execute("INSERT INTO employees (name, salary) VALUES (%s, %s)", ("Gopal", 47000))

    # commit transaction
    con.commit()
    print("Transaction committed successfully")

except:
    # rollback if any error occurs
    con.rollback()
    print("Transaction rolled back due to error")

# display records after transaction
print("\nEmployees after transaction:")
cur.execute("SELECT name, salary FROM employees")
rows = cur.fetchall()
for r in rows:
    print(r[0], "-", r[1])

# close connection
cur.close()
con.close()

```

the salary of an employee, but rolls back the transaction if an error occurs during
the update.**



```python
import mysql.connector

# connect to MySQL database
con = mysql.connector.connect(
    host="localhost",
    user="root",
    password="2910",
    database="college"
)

# buffered cursor to avoid unread result error
cur = con.cursor(buffered=True)

# show data BEFORE update
print("Employee data BEFORE transaction:")
cur.execute("SELECT name, salary FROM employees WHERE name = %s", ("Rohit",))
row = cur.fetchone()
print(row[0], "-", row[1])

try:
    # start transaction
    con.start_transaction()
    print("\nTransaction started")

    # update salary
    cur.execute("UPDATE employees SET salary = %s WHERE name = %s", (70000, "Rohit"))

    # intentional error (wrong column name) to trigger rollback
    cur.execute("UPDATE employees SET salry = %s WHERE name = %s", (75000, "Rohit"))

    # commit if no error
    con.commit()
    print("Transaction committed")

except:
    # rollback on error
    con.rollback()
    print("Error occurred! Transaction rolled back")

# show data AFTER transaction
print("\nEmployee data AFTER transaction:")
cur.execute("SELECT name, salary FROM employees WHERE name = %s", ("Rohit",))
row = cur.fetchone()
print(row[0], "-", row[1])

# close connection
cur.close()
con.close()
```


**7.Write a Python program that connects to a MySQL database and uses a prepared
statement to insert multiple records into a table.**


 

```python
import mysql.connector

# connect to MySQL database
con = mysql.connector.connect(
    host="localhost",
    user="root",
    password="2910",
    database="college"
)

# buffered cursor
cur = con.cursor(buffered=True)

# prepared SQL statement
sql = "INSERT INTO employees (name, salary) VALUES (%s, %s)"

# multiple records to insert
data = [
    ("Harshu", 42000),
    ("Harsha", 46000),
    ("Sudhir", 48000)
]

# execute prepared statement for multiple records
cur.executemany(sql, data)

# save changes
con.commit()

print("Multiple records inserted successfully")

# display inserted data
cur.execute("SELECT name, salary FROM employees")
rows = cur.fetchall()

for r in rows:
    print(r[0], "-", r[1])

# close connection
cur.close()
con.close()
```

named students and courses, establishes a relationship between them, and inserts
sample data**


```python
# Question 9:

import mysql.connector

# connect to MySQL database
con = mysql.connector.connect(
    host="localhost",
    user="root",
    password="2910",
    database="college"
)

cur = con.cursor(buffered=True)

# create students table
cur.execute("CREATE TABLE IF NOT EXISTS students (id INT AUTO_INCREMENT PRIMARY KEY, name VARCHAR(50))")

# create courses table with foreign key
cur.execute("CREATE TABLE IF NOT EXISTS courses (id INT AUTO_INCREMENT PRIMARY KEY, course_name VARCHAR(50), student_id INT, FOREIGN KEY (student_id) REFERENCES students(id))")

# insert sample students
cur.execute("INSERT INTO students (name) VALUES (%s)", ("Dilip",))
cur.execute("INSERT INTO students (name) VALUES (%s)", ("Harsh",))

# insert sample courses
cur.execute("INSERT INTO courses (course_name, student_id) VALUES (%s, %s)", ("Python", 1))
cur.execute("INSERT INTO courses (course_name, student_id) VALUES (%s, %s)", ("Java", 2))

# save changes
con.commit()

print("Students and Courses tables created with relationship and sample data inserted")

# close connection
cur.close()
con.close()


```

on the students and courses tables, and prints the names of students along with the
courses they are enrolled in**

```python
import mysql.connector

# connect to MySQL database
con = mysql.connector.connect(
    host="localhost",
    user="root",
    password="2910",
    database="college"
)

cur = con.cursor(buffered=True)

# display students table
print("Students Table:")
cur.execute("SELECT * FROM students")
rows = cur.fetchall()
for r in rows:
    print(r)

# display courses table
print("\nCourses Table:")
cur.execute("SELECT * FROM courses")
rows = cur.fetchall()
for r in rows:
    print(r)

# display combined data using JOIN
print("\nStudents with their Courses:")
cur.execute("SELECT students.name, courses.course_name FROM students JOIN courses ON students.id = courses.student_id")
rows = cur.fetchall()
for r in rows:
    print(r[0], "-", r[1])

# close connection
cur.close()
con.close()
```
