```python

try:
    file = open("sample.txt", "r")
    content = file.read()
    print("File contents:")
    print(content)
except FileNotFoundError:
    print("Error: File not found.")
finally: #run always not depend on exception occurs or not
    try:
        file.close()
    except NameError: #occures if file is not open 
        pass   #if not handled give like NameError: name 'file' is not defined
    print("finally blocked executed")

```


**7.Create a custom exception class called NegativeNumberError to signal invalid input.**


 

```python

class NegativeNumberError(Exception):
    #occurs if negative number is entered
    pass

try:
    num = int(input("Enter a positive number: "))
    print("Given number:", num)
    if num < 0:
        raise NegativeNumberError("Number cannot be negative.")

except NegativeNumberError as e:
    print("Error:", e)

except ValueError:
    print("Error: Please enter an integer.")

```

file path..**



```python
import os

class InvalidFilePathError(Exception):
    #raises if file path is invalid
    pass

try:
    
    path = input("Enter file path: ")
    path = path.replace("\\", "/")   # to convert \ to /
    #correct path :C:\Users\dilip\OneDrive\Desktop\Desktop\FYMCA\PYTHON\Exception_Handling.pdf

    if not os.path.isfile(path):
        raise InvalidFilePathError("Invalid file path: " + path)

    with open(path, "rb") as f: 
        print("File opened successfully.")
        print("File size:", len(f.read()), "bytes")

except InvalidFilePathError as e:
    print("Error:", e)

except FileNotFoundError:
    print("Error: File not found.")
```


**9.Write a function that raises an InsufficientFundsError if a withdrawal amount exceeds the available balance.**


```python

class InsufficientFundsError(Exception):
    #raises if current balance is less than withdrawal amount
    pass

def withdraw(balance, amount):
    if amount > balance:
        raise InsufficientFundsError("Withdrawal amount exceeds current balance")
    return balance - amount

try:
    bal = float(input("Enter current balance: "))
    amt = float(input("Enter withdrawal amount: "))
    print("Current balance:",bal)
    print("Withdrawal ammount:",amt)
    new_balance = withdraw(bal, amt)
    print("Withdrawal successful. New balance:", new_balance)
except InsufficientFundsError as e:
    print("Error:", e)
except ValueError:
    print("Error: Please enter numeric values.")


```

.**

```python

class InvalidPasswordError(Exception):
    #if user enters incorrect password 3 times
    pass

correct_password = "dilip2910"
attempts = 0

try:
    while attempts < 3:
        password = input("Enter password: ")
        print("entered password:",password)
        if password == correct_password:
            print("Login successful")
            break   
        else:
            attempts += 1
            print("Incorrect password. Attempts left:", 3 - attempts)

    if attempts == 3:  
        raise InvalidPasswordError("entered incorrect password 3 times.")

except InvalidPasswordError as e:
    print("Error:", e)




```
