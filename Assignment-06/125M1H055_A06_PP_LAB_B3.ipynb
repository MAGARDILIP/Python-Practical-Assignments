**1.Write a Python program to list all files and directories in the current working directory
using the os module**


```python
import os

print(f"Showing data of: {os.getcwd()}")

data = os.listdir('.')

files = []       # to store files
directories = [] # to store directories

for entry in data:

    if os.path.isdir(entry):
        directories.append(entry)
 
    elif os.path.isfile(entry):
        files.append(entry)


print("\nDirectories Found:")
for d in directories:
    print(f"{d}")

print("\nFiles Found:")
for f in files:
    print(f"{f}")

```

directory to it, and then removes it using the os module.**


```python

import os  # for getcwd(),mkdir(),chdir(),rmdir()


DIR_NAME = "new_folder"

original_dir = os.getcwd()

print(f"Current working directory: {original_dir}")


print(f"\ncreating new directory with name {DIR_NAME}")
os.mkdir(DIR_NAME)
print(f"\ndirectory with name {DIR_NAME} is created")

print(f"changing CWD to {DIR_NAME}")
os.chdir(DIR_NAME)
print(f"New Current working directory: {os.getcwd()}")


print("\nChanging CWD back to previous original directory...")
os.chdir(original_dir)
print(f"Now CWD : {os.getcwd()}")


print(f"\nRemoving directory: {DIR_NAME}")
os.rmdir(DIR_NAME)
print("Directory removed")

```


**7.Write a Python program to generate a list of prime numbers up to a given number
using the math.isqrt() method**


 

```python
import math

n=int(input("enter a number"))
print(f"Prime Numbers up to {n} ")

primes = []


def is_prime(num):
    if num <= 1:
        return False
    
    max_divisor = math.isqrt(num)
    
    for i in range(2, max_divisor + 1):
        if num % i == 0:
            return False 
    return True

for number in range(2, n + 1):
    if is_prime(number):
        primes.append(number)

print(primes)
```

two decimal places using the math module.**



```python
import math

number = float(input("Enter a non-negative number: "))


sqrt = math.sqrt(number)


result = round(sqrt, 2)

print(f"\nOriginal Number: {number}")
print(f"Square Root: {sqrt}")
print(f"Result after rounding square root to two dedimal places: {result}")
```
