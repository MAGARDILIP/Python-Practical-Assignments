**1.Create a Vehicle class with max_speed and mileage instance attributes and print max_speed and mileage.**


```python
# Question 1: 


class Vehicle:
    def __init__(self, max_speed, mileage):
      # def __init__(self): This method is automatically executed when you create an object.
        self.max_speed = max_speed     # we can also create instance variables outside the class or in another method 
        self.mileage = mileage

# creating an object
car = Vehicle(180, 15000)     # internally converts this into => Vehicle.__init__(car, 180, 15000)

print("Max Speed:", car.max_speed)
print("Mileage:", car.mileage)


# __init__        =>Python’s constructor         Runs automatically when object is created

#  self           =>Refers to the object        Same as Java’s this      Needed to create and access instance variables
#self MUST be there because every instance method receives the object as first argument.

# Python NEVER declares instance variables outside the constructor
#Python does not automatically differentiate instance vs local variables so wealways need to use self 
```

```python
# Question 2:  

class TwoSum:
    def __init__(self, nums):
        self.nums = nums

    def find_indices(self, target):
        hash_map = {}   # Stores: number → index

        for index, num in enumerate(self.nums):
          # enumerate() => It gives both index and value
            secondNumber = target - num   # find needed number to reach target

            if secondNumber in hash_map:
                return (hash_map[secondNumber], index)  # return indices
            
            hash_map[num] = index   # store current number
            
        return None  # no pair found

nums = [2, 7, 11, 15]
print(nums)
target = 9

obj = TwoSum(nums)
result = obj.find_indices(target)

if result is None:
    print(f"There are no numbers whose sum = {target}")
else:
    print(f"Numbers at indices {result} give sum = {target}")
```

**3.Create a Bus child class that inherits from the Vehicle class. The default fare charge of any vehicle is seating capacity * 100. If Vehicle is a Bus instance, we need to add an extra 10% on the full fare as a maintenance charge. So total fare for bus instance will become the final amount = total fare + 10% of the total fare.**


```python
# Question 3:
class Vehicle:
    def __init__(self, capacity):
        self.capacity = capacity

    def fare(self):
        return self.capacity * 100


class Bus(Vehicle):
    def fare(self):
        total_fare = super().fare()      # get fare from Vehicle class
        maintenance_charge = total_fare * 0.10
        return total_fare + maintenance_charge


bus = Bus(50)
print("Bus capacity : ",bus.capacity)
print("Total Bus Fare:", bus.fare())

```

```python
# Question 4: 
class IntegerToRoman:
    def int_to_roman(self, num):
      
        values =    [1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1]
        symbols = ["M", "CM","D","CD","C","XC","L","XL","X","IX","V","IV","I"]

        roman = ""

        # Convert number to roman numeral
        for i in range(len(values)):
            while num >= values[i]:
                roman += symbols[i]
                num -= values[i]

        return roman
      


obj = IntegerToRoman()

print(f"Conversion of 2025 to Roman => {obj.int_to_roman(2025)}")   # MMXXIV
print(f"Conversion of 58 to Roman => {obj.int_to_roman(58)}")     # LVIII
print(f"Conversion of 10 to Roman => {obj.int_to_roman(10)}")   # MCMXCIV

```

```python
# Question 5:
class Validator:
    def is_valid(self, s):
        stack = []
        
        # mapping of closing -> opening brackets
        match = {')': '(', '}': '{', ']': '['}

        for char in s:
            # If it is  opening bracket, push to stack
            if char in "({[":
                stack.append(char)
            
            # If it is closing bracket
            elif char in ")}]":
                # If stack is empty OR top does not match -> invalid
                if not stack or stack[-1] != match[char]:
                    return False
                stack.pop()  # remove the matched opening bracket
        
        # If stack is empty -> all brackets matched
        return len(stack) == 0


obj = Validator()

tests = ["()", "()[]{}", "[)", "((((", ")))))"]

for t in tests:
    print(f"{t} => {obj.is_valid(t)}")


```

```python
# Python supports 5 types of inheritance:

# Single     # Multilevel  # Hierarchical    # Hybrid    # Multiple
 
#python support multiple inheritance unlike java because Python uses
# Method Resolution Order (MRO) and the C3 linearization algorithm to resolve conflicts when a method appears in multiple parents.

# Question 10:

class Animal:
    def make_sound(self):
        print("Animal makes a sound")

class Dog(Animal):
    def make_sound(self):
        print("Dog barks")


class Cat(Animal):
    def make_sound(self):
        print("Cat meows")


a = Animal()
d = Dog()
c = Cat()

# Calling overridden methods
a.make_sound()   # Animal method
d.make_sound()   # Dog method
c.make_sound()   # Cat method


```
