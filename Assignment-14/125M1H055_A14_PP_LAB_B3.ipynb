## What is SQLite?

SQLite is a lightweight, file-based relational database management system (RDBMS) that stores the entire database in a single `.db` file and does not require a separate server to run.

## Why we are using SQLite

- SQLite is easy to use and comes built-in with Python through the `sqlite3` module.
- It does not require a username, password, or server setup, making it suitable for beginners.
- It is lightweight and fast, ideal for small applications and college assignments.
- SQLite supports standard SQL commands like `CREATE`, `INSERT`, `SELECT`, `UPDATE`, and `DELETE`.
- It works perfectly in Jupyter Notebook for database handling programs.

# SQLite Functions Used in the Program

## 1. sqlite3.connect()

**Purpose:**  
The `connect()` function is used to connect Python with an SQLite database.  
If the database file does not exist, SQLite automatically creates a new database file.

**Syntax:**  
sqlite3.connect(database_name)

---

## 2. execute()

**Purpose:**  
The `execute()` function is used to run SQL commands such as creating tables, inserting records, selecting data, updating records, and deleting records.  
In SQLite, basic SQL operations can be executed directly using the connection object without creating a cursor.

**Syntax:**  
connection.execute(sql_query)  
connection.execute(sql_query, values)

---

## 3. commit()

**Purpose:**  
The `commit()` function is used to save the changes permanently in the database.  
It is required after insert, update, and delete operations.

**Syntax:**  
connection.commit()

---

## 4. SELECT Operation

**Purpose:**  
The SELECT statement is used to retrieve all records from a table in the database.

**Syntax:**  
SELECT * FROM table_name

---


## 6. close()

**Purpose:**  
The `close()` function is used to close the database connection and release system resources after all database operations are completed.

**Syntax:**  
connection.close()

---

## 7. Parameter Placeholder (?)

**Purpose:**  
The `?` symbol is used as a placeholder to safely pass values into SQL queries.  
It helps prevent SQL injection and avoids SQL syntax errors.

**Syntax:**  
INSERT INTO table_name VALUES (?, ?)

---

## 5. UPDATE Statement

**Purpose:**  
The UPDATE statement is used to modify the value of a specific column in an existing row of a table.

**Syntax:**  
UPDATE table_name SET column_name = value WHERE condition

---

## UPDATE Statement (Without WHERE Clause)

**Purpose:**  
The UPDATE statement without a WHERE clause is used to modify the values of a specific column for **all rows** in a table.

**Syntax:**  
UPDATE table_name SET column_name = value

---


## 6. DELETE Statement

**Purpose:**  
The DELETE statement is used to remove a specific row from a table based on a given condition.

**Syntax:**  
DELETE FROM table_name WHERE condition

---

## 7. WHERE Clause

**Purpose:**  
The WHERE clause is used to apply a condition so that UPDATE and DELETE operations affect only specific rows instead of the entire table.

**Syntax:**  
WHERE condition

---

## 8. Parameter Placeholder (?)

**Purpose:**  
The `?` placeholder is used to safely pass values into SQL queries and helps prevent SQL injection.

**Syntax:**  
SET column_name = ?  
WHERE column_name = ?

---

## 9. ALTER TABLE Statement

**Purpose:**  
The ALTER TABLE statement is used to modify the structure of an existing table.  
It can be used to add a new column to a table.

**Syntax:**  
ALTER TABLE table_name ADD COLUMN column_name data_type

**1.Write a Python program to create a table and insert some records in that table.
Finally selects all rows from the table and display the records
.**


```python
import sqlite3

# connect to database
con = sqlite3.connect("college.db")

# create table

con.execute("CREATE TABLE IF NOT EXISTS student (id INTEGER PRIMARY KEY, name TEXT, marks INTEGER)")


print("Table created")

# insert records
con.execute("INSERT INTO student (name, marks) VALUES (?, ?)", ("Dilip", 85))
con.execute("INSERT INTO student (name, marks) VALUES (?, ?)", ("Harsh", 78))
con.execute("INSERT INTO student (name, marks) VALUES (?, ?)", ("Rohit", 90))

print("Records inserted")

# select and display records
rows = con.execute("SELECT * FROM student")

print("Student Records:")
for r in rows:
    print(r)

# save changes
con.commit()

# close connection
con.close()
```



**2.Write a Python program to update a specific column value of a given table and
select all rows before and after updating the said table.**



```python
# Question 2:  
import sqlite3

# connect to database
con = sqlite3.connect("college.db")

# display records before update
print("Records BEFORE update:")
rows = con.execute("SELECT * FROM student")
for r in rows:
    print(r)

# update specific column value
con.execute("UPDATE student SET marks = ? WHERE name = ?",(95, "Dilip"))

# save changes
con.commit()

# display records after update
print("\nRecords AFTER update:")
rows = con.execute("SELECT * FROM student")
for r in rows:
    print(r)

# close connection
con.close()


```

```python
# Question 3:

import sqlite3

# connect to database
con = sqlite3.connect("college.db")

# display records before deletion
print("Records BEFORE deletion:")
rows = con.execute("SELECT * FROM student")
for r in rows:
    print(r)

# delete a specific row
con.execute(
    "DELETE FROM student WHERE name = ?",
    ("Harsh",)
)

# save changes
con.commit()

# display records after deletion
print("\nRecords AFTER deletion:")
rows = con.execute("SELECT * FROM student")
for r in rows:
    print(r)

# close connection
con.close()
```

## Why MySQL is used with Python

**6.Write Python code to:**

**a. Connect to a MySQL database using the mysql.connector library.**

**b. Execute a SQL query to retrieve all customer names and emails.**

**c. Insert a new row into the "customers" table.**

**d. Update an existing customer's email address**

```python
import mysql.connector

# connect to MySQL server
con = mysql.connector.connect(
    host="localhost",
    user="root",
    password="2910"
)

print("Connected to MySQL Server")

cur = con.cursor()

# create database
cur.execute("CREATE DATABASE IF NOT EXISTS college")
print("Database college created or already exists")

# use database
cur.execute("USE college")

# create customers table
cur.execute("""
CREATE TABLE IF NOT EXISTS customers (
    id INT AUTO_INCREMENT PRIMARY KEY,
    name VARCHAR(50),
    email VARCHAR(50)
)
""")

print("Customers table created or already exists")

# save changes
con.commit()

# close connection
cur.close()
con.close()



```

```python
import mysql.connector

# connect to MySQL database
con = mysql.connector.connect(
    host="localhost",
    user="root",
    password="2910",
    database="college"
)

print("Connected to MySQL")

cur = con.cursor()

# a) retrieve all customer names and emails
print("\nCustomer names and emails:")
cur.execute("SELECT name, email FROM customers")
rows = cur.fetchall()

for r in rows:
    print(r)

# b) insert a new row into customers table
cur.execute(
    "INSERT INTO customers (name, email) VALUES (%s, %s)",
    ("Dilip", "dilip@gmail.com")
)

print("\nNew customer inserted")

# c) update an existing customer's email
cur.execute(
    "UPDATE customers SET email = %s WHERE name = %s",
    ("dilip_new@gmail.com", "Dilip")
)

print("Customer email updated")

# save changes
con.commit()

# close connection
cur.close()
con.close()
```


**7.Write a Python program that connects to a MySQL database and fetches and
prints the version of the MySQL server.**


 

```python
# Question 7:

import mysql.connector

# connect to MySQL database
con = mysql.connector.connect(
    host="localhost",
    user="root",
    password="2910"
)

cur = con.cursor()

# fetch MySQL server version
cur.execute("SELECT VERSION()")
version = cur.fetchone()

print("MySQL Server Version:", version[0])
#The fetchone() method returns a tuple, so [0] is used to access the actual value from the tuple.

# close connection
cur.close()
con.close()
```

the databases available on the server..**



```python
import mysql.connector

# connect to MySQL server
con = mysql.connector.connect(
    host="localhost",
    user="root",
    password="2910"
)

cur = con.cursor()

# list all databases
cur.execute("SHOW DATABASES")
dbs = cur.fetchall()

print("Databases available on the server:")
for db in dbs:
    print(db[0])

# close connection
cur.close()
con.close()

```

named employees with columns for id, name, and salary, and inserts a few
records into the table.**


```python
import mysql.connector

# connect to MySQL database
con = mysql.connector.connect(
    host="localhost",
    user="root",
    password="2910",
    database="college"
)

cur = con.cursor()

# create employees table (single-line query)
cur.execute("CREATE TABLE IF NOT EXISTS employees (id INT AUTO_INCREMENT PRIMARY KEY, name VARCHAR(50), salary INT)")

print("Employees table created")

# insert records (single-line queries)
cur.execute("INSERT INTO employees (name, salary) VALUES (%s, %s)", ("Dilip", 50000))
cur.execute("INSERT INTO employees (name, salary) VALUES (%s, %s)", ("Harsh", 45000))
cur.execute("INSERT INTO employees (name, salary) VALUES (%s, %s)", ("Rohit", 55000))

# save changes
con.commit()

print("Records inserted successfully")

print("Employees data:")

# display data
cur.execute("SELECT * FROM employees")
rows = cur.fetchall()

for r in rows:
    print(r)

# close connection
cur.close()
con.close()
```

```python
import mysql.connector

# connect to MySQL database
con = mysql.connector.connect(
    host="localhost",
    user="root",
    password="2910",
    database="college"
)

cur = con.cursor()

# fetch names and salaries
cur.execute("SELECT name, salary FROM employees")
rows = cur.fetchall()

print("Employee Name and Salary:")

for r in rows:
    print(r[0], "-", r[1])

# close connection
cur.close()
con.close()
```
