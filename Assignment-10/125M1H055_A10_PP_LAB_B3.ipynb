```python
data = []

try:
    with open("sample.txt", "r") as file:
        for line in file:
            data.append(line)   #without using strip()
    print("data in sample.txt without using strip()\n")
    for item in data:
        print(item)  

except FileNotFoundError:
    print("Error: sample.txt not found.")
```

```python
data = []

try:
    with open("sample.txt", "r") as file:
        for line in file:
            data.append(line.strip())   # using strip()
#strip() is used to remove extra spaces and the newline \n at the start or end of a line.
    print("data in sample.txt using strip()\n")
    for item in data:
        print(item)

except FileNotFoundError:
    print("Error: sample.txt not found.")


```

```python
try:
    with open("sample.txt", "r") as file:
        content = file.read()                
        print("Content in sample.txt:\n",content)

        words = content.split()
              

    print("After using split(), words list contains:\n", words)

    max_len = max(len(word) for word in words)
    
    #here first we find the length of each word in list and apply max() on it and store max length
    
    longest_words = [word for word in words if len(word) == max_len]
    #check length of each word if it equal to max_len then we store it in word
    print("\nLongest words:", longest_words)

except FileNotFoundError:
    print("Error: The file 'sample.txt' was not found.")
```

```python
import string

n = int(input("Enter number of letters per line: "))
print(f"user wants {n} letters on each line of file")
letters = string.ascii_lowercase    # abcdefghijklmnopqrstuvwxyz

with open("sample2.txt", "w") as file:
    for i in range(0, len(letters), n):
        
        # here i takes values: 0, n, 2n, 3n, ...
        #assume n=4
        #here i can have values 0 4 8 12 16 20 24
        
        file.write(letters[i:i+n] + "\n")
        
        #means on first line of file letters[0:4] i.e abcd
        #on second line of file letters[4:8] i.e efgh 

print("sample2.txt file created successfully")
with open("sample2.txt", "r") as file:
    print("Content of sample2.txt:\n")
    print(file.read())

```


**6.Implement a program that opens a file in write mode, but only if it doesn't already
exist**




```python
import os

print("\nFiles in current directory:")
for file in os.listdir():
    print(file)
    
filename = "newfile.txt"

if not os.path.exists(filename):
    with open(filename, "x") as file:
        file.write("\nFile created successfully")
    print("File created:", filename)
else:
    print("File already exists")

```


**7.Write code to read a CSV file containing product information and create a list of product objects**


 

```python
#to read csv file we need csv file so we create it
import csv

data = [
    ["ID", "Name", "Price"],    
    ["201", "Tablet", "15000"],
    ["202", "Charger", "600"],
    ["203", "HDMI Cable", "350"]
]

with open("products.csv", "w", newline="") as file:
    #newline="" prevents writing extra new line into file
    writer = csv.writer(file)
    writer.writerows(data)

print("products.csv created successfully")

```

```python
import csv

product_list = []

try:
    with open("products.csv", "r") as file:
        reader = csv.reader(file)

        for row in reader:
            product_list.append(row)   

    print("Data in products.csv:\n")
    for product in product_list:
        print(product)

except FileNotFoundError:
    print("Error: products.csv not found.")
```



**8.Create a program that generates a CSV report from a list of dictionaries..**



```python
import csv

products = [
    {"ID": "201", "Name": "Tablet", "Price": "15000"},
    {"ID": "202", "Name": "Charger", "Price": "600"},
    {"ID": "203", "Name": "HDMI Cable", "Price": "350"}
]

with open("report.csv", "w", newline="") as file:
    writer = csv.DictWriter(file, fieldnames=products[0].keys())
    # DictWriter is CSV writer used for dictionaries
    # products[0].keys() gives ['ID', 'Name', 'Price']

    writer.writeheader()        # writes: ID, Name, Price
    writer.writerows(products)  # writes product rows

print("report.csv generated\n")


print("Content of report.csv\n")

with open("report.csv", "r") as file:
    print(file.read())
```

```python
try:
    with open("sample.txt", "r") as file:
        content = file.read()
    print(content)
    lines = content.count("\n") + 1 
    # add 1 because no. of lines = no. of newline characters + 1
    #last line does not contain \n character
    words = len(content.split())
    chars = len(content)

    print("Number of lines:", lines)
    print("Number of words:", words)
    print("Number of characters:", chars)

except FileNotFoundError:
    print("Error: sample.txt not found.")

```

```python
filename = "notes.txt"

with open(filename, "w") as file:
    print("Enter lines to write to the file. Enter 0 to stop.\n")

    while True:
        line = input("Enter a line: ")
        if line == "0":     
            break
        file.write(line + "\n")   # write each line on a new line

print("Data written to", filename)


print("Data of notes.txt:\n")

with open(filename, "r") as file:
    print(file.read())
```
