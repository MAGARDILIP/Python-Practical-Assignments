```python
import math

radius = float(input("Enter radius: "))
area = round(math.pi * radius * radius,2)
circumference = round(2 * math.pi * radius, 2)

print("Radius:", radius)
print("Area:", area)
print("Circumference:", circumference)

```


**7.Program to Create a Directory and List Files in it(Using the os module)**


 

```python
import os

dirname = "myfolder2"


if not os.path.exists(dirname):
    os.mkdir(dirname)
    print("Directory created:", dirname)

    files = ["file1.txt", "file2.txt", "file3.txt"]

    for filename in files:
        file_path = os.path.join(dirname, filename)
        open(file_path, "w").close()  
        print(f"{filename} created inside directory: {dirname}")

else:
    print("Directory already exists.")

print("Files in", dirname, ":")
print(os.listdir(dirname))
```

 (Create a module math_operations.py with arithmetic functions)
.**

```python
%%writefile math_umodule.py
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    return a / b
```

```python
# we first create   math_umodule.py file in same folder and import that module

import math_umodule

a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

print("First number:",a)
print("Second number:",b)


print("Addition:", math_umodule.add(a, b))
print("Subtraction:", math_umodule.subtract(a, b))
print("Multiplication:", math_umodule.multiply(a, b))

if b != 0:
    print("Division:", math_umodule.divide(a, b))
else:
    print("Division: Cannot divide by zero")



```
