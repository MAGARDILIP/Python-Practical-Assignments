within multiple layers of function calls..**


```python
try:
    try:
        num1 = int(input("Enter first number: "))   
        num2 = int(input("Enter second number: "))  

        try:
            print(f"Result of {num1} / {num2} =", num1 / num2) 
        except ZeroDivisionError:
            print("Inner: Cannot divide by zero.")
            raise         # sends the  ZeroDivisionError to the outer block.

    except ValueError:
        print("Outer: Invalid number.")
    except ZeroDivisionError:
        print("Outer: Zero division error ")

except Exception as e:   #catches exceptions that not handle by inner or outer blocks
    print("Top level:", e)
```

regardless of exceptions or successful execution.**



```python
file = None

try:
    file_name = input("Enter file name to open: ")
    print(file_name)
    file = open(file_name, "r")
    print("File contents:")
    print(file.read())
except FileNotFoundError:
    print("Error: File not found.")
finally:
    if file is not None:
        file.close()
        print("File closed")
    print("finally block executed.")

```

operations in case of temporary failures**


```python
correct_password = "dilip2910"
max_attempts = 3
attempt = 0

while attempt < max_attempts:
    try:
        print("Attempt:", attempt + 1)
        password = input("Enter password: ")
        print("entered password:",password)

       
        if password != correct_password:
            raise ValueError("Incorrect password. Please try again.")

        print("Login successful")
        break  

    except ValueError as e:
        print(e)
        attempt += 1

if attempt == max_attempts:
    print("Login failed after", max_attempts, "attempts.")
```

mathematical operation (e.g., addition, subtraction, multiplication, division) based on
user input Handle the potential exceptions that may arise during user input or
calculation.**


```python
try:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))
    op = input("Enter operation (+, -, *, /): ")
    
    print(f"evaluating: {num1}{op}{num2}")
    if op == "+":
        result = num1 + num2
    elif op == "-":
        result = num1 - num2
    elif op == "*":
        result = num1 * num2
    elif op == "/":
        result = num1 / num2     
    else:
        raise ValueError("Invalid operation.")

    print("Result:", result)

except ValueError as e:
    print("ValueError:", e)
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
```

password. Handle the KeyError when checking if the entered username exists in a
dictionary and handle the ValueError when comparing the entered password..**

```python

users = {
    "dilip": 1007,
    "harsh": 2910,
    "admin": 9999,
    "guest": 1111
}

try:
    username = input("Enter username: ")
    print("given username",username)
    # May raise KeyError
    correct_pin = users[username]

    password = input("Enter numeric PIN password: ")
    print("given password",password)
     
    # May raise ValueError 
    entered_pin = int(password)

    if entered_pin == correct_pin:
        print("Login successful")
    else:
        print("Incorrect password.")

except KeyError:
    print("Error: Username does not exist.")
except ValueError:
    print("Error: Password must be a number (integer).")


```
