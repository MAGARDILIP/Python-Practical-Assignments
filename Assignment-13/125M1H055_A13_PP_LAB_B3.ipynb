**1.Create a program that models a basic shape hierarchy. Define a base class Shape
with methods for calculating area and perimeter. Create subclasses for Circle and
Rectangle..**


````python
# A dunder method is a method that Starts and ends with double underscores
#ex) __init__ , __del__
# Question 1: 
import math

class Shape:
    def area(self):
        return 0
    
    def perimeter(self):
        return 0


class Circle(Shape): # Circle inheriting from Shape
    def __init__(self, radius):
        self.radius = radius

    # Overriding area method
    def area(self):
        # Area = pi * r^2  
        return math.pi * (self.radius ** 2)

    # Overriding perimeter method
    def perimeter(self):
       # Perimeter = 2 * pi * r
        return 2 * math.pi * self.radius



class Rectangle(Shape):  #  Rectangle inheriting from Shape
    def __init__(self, length, width):
        self.length = length
        self.width = width

    # Overriding area method
    def area(self):
         # Area = length * width
        return self.length * self.width

    # Overriding perimeter method
    def perimeter(self):
        # Perimeter = 2 * (length + width)
        return 2 * (self.length + self.width)


c = Circle(10)
print("Circle:")
print(f"Area = {c.area():.2f}")       
print(f"Perimeter = {c.perimeter():.2f}\n")



r = Rectangle(10, 20)
print("Rectangle:")
print(f"Area = {r.area():.2f}")
print(f"Perimeter = {r.perimeter():.2f}")

````

with a method draw(). Create subclasses Circle and Square that implement the
draw() method.**





##  **What is Polymorphism?**
**Polymorphism = same function name, different behavior depending on the object.**

Example:  
Different classes implement the same method name (`draw()`, `sound()`, etc.) but behave differently.

---

##  Python Supports Polymorphism Through:

### **1️ Method Overriding**
Child class provides its own implementation of a method already defined in the parent class.

### **2️ Operator Overloading**
Operators like `+`, `-`, `*`, `==` can behave differently depending on the object type.  
(Implemented using dunder methods like `__add__`, `__eq__`, etc.)

### **3️ Duck Typing**
"If it walks like a duck and quacks like a duck, Python treats it like a duck."  
Python focuses on **behavior**, not the object's class type.

---

## **Python Does NOT Support Method Overloading**

Example:

```python
def add(x, y):
    return x + y

def add(x, y, z):
    return x + y + z
```

The second definition REPLACES (overwrites) the first one 
Because Python functions are stored by name, and the latest function definition overwrites the previous one 

## BUT Python supports overloading-like behavior using:
### 1. Default parameter
### 2. *args 
### 3. **kwargs 


````python
# Question 2:  


class Shape:
    def draw(self):
        print("Draw() method of Shape class is called ")


# Circle overriding draw()
class Circle(Shape):
    def draw(self):
        print("Draw() method of Circle class is called ")


#  Square overriding draw()
class Square(Shape):
    def draw(self):
        print("Draw() method of Square class is called ")

circle_obj = Circle()
square_obj = Square()

# Calling the same method using different objects
circle_obj.draw()   # Circle
square_obj.draw()   # Square
````

shopping cart. Define a base class Item and subclasses for different types of items.**


````python
# Parent class
class Item:
    def __init__(self, name, base_price, quantity):
        self.name = name
        self.base_price = base_price
        self.quantity = quantity

    def get_total_price(self):
        return self.base_price * self.quantity


# Child class 1: Book (discount)
class Book(Item):
    def __init__(self, name, base_price, quantity, discount):
        super().__init__(name, base_price, quantity)
        self.discount = discount

    def get_total_price(self):
        base_amount = super().get_total_price()
        return base_amount - (base_amount * self.discount)


# Child class 2: Electronic (tax)
class Electronic(Item):
    def __init__(self, name, base_price, quantity, tax):
        super().__init__(name, base_price, quantity)
        self.tax = tax

    def get_total_price(self):
        base_amount = super().get_total_price()
        return base_amount + (base_amount * self.tax)


# Creating a shopping cart with different items
cart = [
    Book("Python Book", 500, 2, 0.10),
    Item("Notebook", 50, 5),
    Electronic("Calculator", 800, 1, 0.18)
]

total_cost = 0

print("Student Shopping Cart Details\n")
print("-" * 45)

for item in cart:
    base_amount = item.base_price * item.quantity
    final_price = item.get_total_price()

    print(f"Item Name  : {item.name}")
    print(f"Base Price : ₹{item.base_price}")
    print(f"Quantity   : {item.quantity}")
    print(f"Base Amt   : ₹{base_amount}")

# isinstance(object, class_or_tuple)  
#Checks whether an object belongs to a given class or any of its parent classes. 
# It checks:
# What class was used to create item?
# Does that class match Book?
# Or does it inherit from Book? 
# True -> if object is of that class
# False -> otherwise

    if isinstance(item, Book):
        discount_amt = base_amount * item.discount
        print(f"Discount   : -₹{discount_amt:.2f}")

    if isinstance(item, Electronic):
        tax_amt = base_amount * item.tax
        print(f"Tax        : +₹{tax_amt:.2f}")

    print(f"Final Amt  : ₹{final_price:.2f}")
    print("-" * 45)

    total_cost += final_price

print(f"\nTotal Cost of items in shopping cart: ₹{total_cost:.2f}")
````

methods for calculating area and perimeter. Create subclasses Circle and
Rectangle to implement these methods..**



````python
# Question 4: 
#    It is not compulsory to write pass, but it is used as a placeholder because Python does not allow empty method bodies.An abstract class is a class that cannot be instantiated and is used to define a common interface for its subclasses. Abstract classes are created using ABC

# They define what methods must exist, not how they work to create an abstract class in Python?
# Steps:

# Import ABC

# Inherit from ABC  to make methods abstract?
# Use @abstractmethod decorator  ... (ellipsis) is a valid Python statement

# It means: “implementation will be provided later”   

# Import ABC and abstractmethod to create abstract classes
from abc import ABC, abstractmethod

# Import math module to use mathematical constants like pi
import math


# Abstract Base Class
class Shape(ABC):
    """
    Shape is an abstract class.
    It defines a common structure for all shapes.
    """

    @abstractmethod
    def area(self):
        # Abstract method: must be implemented by child classes
        pass

    @abstractmethod
    def perimeter(self):
        # Abstract method: must be implemented by child classes
        pass


# Concrete class Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        # Using math.pi for accurate value of π
        return math.pi * self.radius * self.radius

    def perimeter(self):
        return 2 * math.pi * self.radius


# Concrete class Rectangle
class Rectangle(Shape):
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def area(self):
        return self.length * self.breadth

    def perimeter(self):
        return 2 * (self.length + self.breadth)


# Main Program
shapes = [
    Circle(5),
    Rectangle(4, 6)
]

for shape in shapes:
    print("Area:", shape.area())
    print("Perimeter:", shape.perimeter())
    print("----------------------")

````

payments. Implement the interface in classes CreditCardPayment and
PayPalPayment**


**NOTE**

**Python does NOT have a separate interface concept like Java.**

**What we call an “interface” in Python is actually a special case of an abstract class.**

````python
# Question 5:
from abc import ABC, abstractmethod

# Interface
class Payment(ABC):

    @abstractmethod
    def process_payment(self, amount):
        pass


# Credit Card Payment
class CreditCardPayment(Payment):

    def process_payment(self, amount):
        print(f"Processing credit card payment of ₹{amount}")


# PayPal Payment
class PayPalPayment(Payment):

    def process_payment(self, amount):
        print(f"Processing PayPal payment of ₹{amount}")


# Direct object creation and method calls
cp = CreditCardPayment()
cp.process_payment(1500)

pp = PayPalPayment()
pp.process_payment(2500)
````


**6.Write a Python program that models a simple bank account hierarchy. Create a
base class Account with attributes for account_number, account_holder, and
balance. Create subclasses for SavingsAccount and CurrentAccount..**




````python
# Base class Account
class Account:
    def __init__(self, acc_no, name, bal):
        self.acc_no = acc_no      # account number
        self.name = name          # account holder name
        self.bal = bal            # balance

    def show(self):
        print("Account No:", self.acc_no)
        print("Name:", self.name)
        print("Balance:", self.bal)


# Savings Account class
class SavingsAccount(Account):
    def __init__(self, acc_no, name, bal, rate):
        super().__init__(acc_no, name, bal)   # calling parent constructor
        self.rate = rate                      # interest rate

    def show(self):
        super().show()        # calling parent show method
        print("Interest Rate:", self.rate)


# Current Account class
class CurrentAccount(Account):
    def __init__(self, acc_no, name, bal, limit):
        super().__init__(acc_no, name, bal)   # calling parent constructor
        self.limit = limit                    # overdraft limit

    def show(self):
        super().show()        # calling parent show method
        print("Overdraft Limit:", self.limit)


# creating objects
s = SavingsAccount(101, "Rahul", 5000, 4)
c = CurrentAccount(201, "Amit", 8000, 10000)

print("Savings Account")
s.show()

print("\nCurrent Account")
c.show()

````

for name and sound. Create subclasses for different types of animals (e.g., Lion,
Elephant) that inherit from the Animal class..**


 

````python
# Base class Animal
class Animal:
    def __init__(self, name, sound):
        self.name = name      # animal name
        self.sound = sound    # animal sound

    def show(self):
        print("Animal Name:", self.name)
        print("Sound:", self.sound)


# Lion class
class Lion(Animal):
    def __init__(self, name):
        super().__init__(name, "Roar")   # calling parent constructor


# Elephant class
class Elephant(Animal):
    def __init__(self, name):
        super().__init__(name, "Trumpet")   # calling parent constructor


# creating objects
l = Lion("Lion")
e = Elephant("Elephant")

print("Lion Details")
l.show()

print("\nElephant Details")
e.show()

````

Vehicle with methods for start_engine() and stop_engine(). Create subclasses for
Car and Motorcycle that implement these methods.**



````python
# Base class Vehicle
class Vehicle:
    def start_engine(self):
        print("Engine started")

    def stop_engine(self):
        print("Engine stopped")


# Car class
class Car(Vehicle):
    def start_engine(self):
        print("Car engine started")

    def stop_engine(self):
        print("Car engine stopped")


# Motorcycle class
class Motorcycle(Vehicle):
    def start_engine(self):
        print("Motorcycle engine started")

    def stop_engine(self):
        print("Motorcycle engine stopped")


# creating objects
c = Car()
m = Motorcycle()

print("Car")
c.start_engine()
c.stop_engine()

print("\nMotorcycle")
m.start_engine()
m.stop_engine()
````

Define a base class Course with methods for enrolling and dropping students.
Create subclasses for different types of courses (e.g., MathCourse,
HistoryCourse).**


````python
# Base class Course
class Course:
    def __init__(self, name):
        self.name = name
        self.students = []     # list to store students

    def enroll(self, s):
        self.students.append(s)
        print(s, "enrolled in", self.name)

    def drop(self, s):
        if s in self.students:
            self.students.remove(s)
            print(s, "dropped from", self.name)
        else:
            print(s, "not found in", self.name)


# Math course class
class MathCourse(Course):
    def __init__(self):
        super().__init__("Math")


# History course class
class HistoryCourse(Course):
    def __init__(self):
        super().__init__("History")


# creating objects
m = MathCourse()
h = HistoryCourse()

m.enroll("Dilip")
m.enroll("Harsh")

h.enroll("Sudhir")

m.drop("Dilip")
h.drop("Gopal")  # not found

````
