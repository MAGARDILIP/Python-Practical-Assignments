**Question 1: Python Program to Find Numbers Divisible by Another Number.**

```python
def find_div_nums(num_list, divisor):
    new_list = [num for num in num_list if num % divisor == 0]
    #new_list stores numbers which are divisible by divisor from num_list
    return new_list


num_list = []

n= int(input("enter no elements to add in list"))

for i in range(0,n):
    num=int(input(f"enter element {i}"))
    num_list.append(num)  #add numbers in list
    

divisor=int(input("enter divisor:"))

print("Given list",num_list)
print(f"numbers from list divisible by {divisor} are:{find_div_nums(num_list,divisor)}")
```

 **Question 4: Python Program to Find the H.C.F and L.C.M of Two input Numbers**

```python
def cal_hcf(x, y):
   
    
    smaller = min(x, y) #as a number cannt have  factor greater than itself
    hcf = 1             #minimum possible HCF
    for i in range(1, smaller + 1):  #including the smaller number.
        if (x % i == 0) and (y % i == 0):
            hcf = i      # i is a common factor
    return hcf

def cal_lcm(x, y):

    greater = max(x, y)
    
    lcm = greater  # LCM must be at least the greater number.
    
    # WE CAN ALSO USE THIS TO FIND :LCM = (n1*n2)/HCF
    
    while True:    #until LCM found
        if (lcm % x == 0) and (lcm % y == 0):
            return lcm
        lcm += greater

n1 = int(input("enter a number 1 to find LCM and HCF"))
n2 = int(input("enter a number 2 to find LCM and HCF"))
hcf_result = cal_hcf(n1, n2)
lcm_result = cal_lcm(n1, n2)

print(f"The H.C.F. of {n1} and {n2} is: {hcf_result}")
print(f"The L.C.M. of {n1} and {n2} is: {lcm_result}")
```

**Question 5:Python Program to Find Factors of Number**

```python
def factors(number):
    if number <= 0:
        return "enter positive integer."
        
    factors = []
   
    for i in range(1, number + 1):  #Factors include 1 and number itself
        if number % i == 0:
            factors.append(i)
    return f"The factors of {number} are: {factors}"

n =int(input("enter a number to find its factor:"))
print(factors(n))
```

**Question 7:Python Program to shuffle a deck of cards using the module random and draw 5 Cards**

```python
import random

def deck():
   
    suits = ['Hearts', 'Diamonds', 'Clubs', 'Spades']
    ranks = ['2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King', 'Ace']
    deck = [rank + ' of ' + suit for suit in suits for rank in ranks]
    
    # deck = []
    # for suit in suits:
    #     for rank in ranks:
    #         card_name = rank + ' of ' + suit
    #         deck.append(card_name)  
    # return deck

    return deck

def shuffle_and_draw(deck, cards_to_draw=5):

    shuffled_deck = deck[:] #Make a copy to keep original as it is
    random.shuffle(shuffled_deck)
    
    drawn_cards = shuffled_deck[:cards_to_draw]
    return drawn_cards

full_deck = deck()
drawn_cards = shuffle_and_draw(full_deck, 5)

print(f"Total cards in the deck: {len(full_deck)}")
print("\nYour drawn 5 cards:")
for card in drawn_cards:
    print(f" {card}")
```

```python
import calendar

def display(year, month):
    if year<0 or month>12 or month < 0:
        return "please provide valid year and month"
    
    cal = calendar.month(year, month)
    
    # month_name is list which contain all month names 
    print(f"\nCalendar for {calendar.month_name[month]} {year}")
    print("-" * 35)
    print(cal)


year = int(input("Please enter the Year >0 (e.g., 2025): "))

month = int(input("Please enter the Month > 0 and < 13(1=Jan, 12=Dec): "))


display(year, month)
```

**Question 9:Python Program to fnd sum of natural numbers up to n using recursive function**

```python
def sum(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return n + sum(n - 1)

n = int(input("enter value of n"))
result = sum(n)
print(f"The sum of natural numbers up to {n} is: {result}")
```
